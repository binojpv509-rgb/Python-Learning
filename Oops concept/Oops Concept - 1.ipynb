```python
class student:
    def __init__(self, name, age, grade, addr, phone):
        self.name = name
        self.age = age
        self.grade = grade
        self.addr = addr
        self.phone = phone
    def greet(self):
        print(f"Hello, my name is {self.name}. I am {self.age} years old and in grade {self.grade}.")
    def increase_age_10_years(self):
        self.age += 10        
```

```python
student1 = student("Alice", 15, 10, "123 Main St", "555-1234")
student1.greet()
```

```python
student1.increase_age_10_years()
student1.greet()
```

```python
class ticket:
    def __init__ (self, name, age, movie, num_tickets):
        self.name = name
        self.age = age
        self.movie = movie
        self.num_tickets = num_tickets
    def calculate_total_price (self):
        if(self.age < 10):
            price_per_ticket = 100
        elif(self.age >= 10 and self.age <= 60):
            price_per_ticket = 150
        else:
            price_per_ticket = 120
        total_price = price_per_ticket * self.num_tickets
        return total_price
    def book_tickets(self):
        total_price = self.calculate_total_price()
        print(f"Booking {self.num_tickets} tickets for {self.movie}. Total price is {total_price}.")
        
if __name__ == "__main__":
    name = input("Enter your name: ")
    age = int(input("Enter your age: "))
    movie = input("Enter movie name: ")
    num_tickets = int(input("Enter number of tickets: "))
    
    ticket_booking = ticket(name, age, movie, num_tickets)
    ticket_booking.book_tickets()
```
